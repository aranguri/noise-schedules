```python
from utils import load_low
import numpy as np
import matplotlib.pyplot as plt
```

```python
vs = ['vp', 've']
lbs = {'vp': 'VP SDE', 've': 'VE SDE'}
ns = {'vp': [10, 25, 50, 100, 250, 333, 500, 1000], 've': [333, 500, 750, 1000]}

plt.figure(figsize=(6, 4), dpi=300)  # Adjust the figure size and resolution
for v in vs:
    ps = []
    for n in ns[v]:
        ls = load_low(v, n)[:7500]
        print(len(ls))
        ps.append(np.mean(ls))
    plt.plot(ns[v], (1-np.array(ps))*100, marker='o', label=lbs[v])
plt.title('Low-level feature', fontsize=14)
plt.xlabel('Discretization steps', fontsize=12)
plt.ylabel('% of unrecognized images', fontsize=12)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.savefig('low.png', bbox_inches='tight')
```
