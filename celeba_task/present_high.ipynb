```python
import numpy as np
import matplotlib.pyplot as plt
from utils import load_high

nums = {'vp': [10, 25, 50, 100, 250, 333, 500, 1000], 've': [333, 500, 750, 1000], 'og': [0]}
#nums = {'ve': [300, 450, 650, 800, 1000], 'og': [0]}

def data(var, n):
    vs = np.array([np.argmax(list(p.values())) for p in load_high(var, n) if p is not None])[:7500]
    if var == 'vp' or var == 've': print(len(vs))
    ps = np.array([(vs == i).mean() for i in range(6)])
    return vs, ps

def present(var):
    for n in nums[var]: print(data(var, n)[1])

kl  = lambda p, q: (p * np.log(p/q)).sum()
kls = lambda var: [kl(data(var, n)[1], data('og', 0)[1]) for n in nums[var]]
```

```python
ve = kls('ve')
```

```python
vp = kls('vp')
```

```python
plt.figure(figsize=(6, 4), dpi=300)  # Adjust the figure size and resolution
plt.title('High-level feature', fontsize=14)
plt.plot(nums['vp'], vp, marker='o', label='VP SDE')
plt.plot(nums['ve'], ve, marker='o', label='VE SDE')

plt.xlabel('Discretization steps', fontsize=12)
plt.ylabel('KL', fontsize=12)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.savefig('high.png', bbox_inches='tight')
```
